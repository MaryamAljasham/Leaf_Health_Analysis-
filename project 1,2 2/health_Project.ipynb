```python
q1=df.quantile(q=.25)
q2=df.quantile(q=.75)
IQR=(q2-q1)
lower=q1-1.5*IQR
upper=q2+1.5*IQR 
df_clean = df[~((df.Value < lower.Value) | (df.Value > upper.Value))]
df_clean
```

```python
def outliers(col):
    list_row=df[col].unique().tolist()
    df_clean=()
    total={}
    for i in list_row:
        grouped = df.groupby(df[col])
        df_new = grouped.get_group(i)
        de=df_new.describe()
        IQR=(de.iloc[6]-de.iloc[4])
        lower=de.iloc[4]-1.5*IQR
        upper=de.iloc[6]+1.5*IQR 
        mask =len(df_new[(df_new.Value >= upper.Value)|(df_new.Value <= lower.Value)])
        total[i]=[mask]
         
    return total
```
